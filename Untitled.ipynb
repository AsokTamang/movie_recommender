**Calculation of cost function from scratch**

```python
def cofi_cost_func(X, W, b, Y, R, lambda_):
    nm, nu = Y.shape  #nm is the number of movies and nu is the number of users
    J = 0
    for j in range(nu):
        for i in range(nm):
            if R[i,j] == 1:  #if the movie is rated by the current user, then only we do the calculation of the cost function for the movie
                J += np.sum(((W[j]@X[i]) + b[0,j] - Y[i,j]) ** 2)
    J = J/2
    regularization = ((lambda_/2) * (np.sum(W**2))) +  (((lambda_/2) * np.sum(X **2)))  #we also use regularization to prevent the overfitting of the model on the training dataset
    J = J + regularization
    return J
```
