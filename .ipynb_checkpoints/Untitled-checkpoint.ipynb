```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from utils import *
```

**Loading the dataset**

```python
X, W, b, num_movies, num_features, num_users = load_precalc_params_small()
Y, R = load_ratings_small() #here Y is the actual rating value of all the movies rated by all the users, R represents whether the movie is rated or not rated

print("Y", Y.shape, "R", R.shape)
print("X", X.shape)
print("W", W.shape)
print("b", b.shape)
print("num_features", num_features)
print("num_movies",   num_movies)
print("num_users",    num_users)
```

**Calculation of cost function from scratch**

```python
def cofi_cost_func(X, W, b, Y, R, lambda_):
    nm, nu = Y.shape  #nm is the number of movies and nu is the number of users as movies are arranged as rows and the users are arranged as columns in case of y
    J = 0
    for j in range(nu):
        for i in range(nm):
            if R[i,j] == 1:  #if the movie is rated by the current user, then only we do the calculation of the cost function for the movie
                J += np.sum(((W[j]@X[i]) + b[0,j] - Y[i,j]) ** 2)
    J = J/2
    regularization = ((lambda_/2) * (np.sum(W**2))) +  (((lambda_/2) * np.sum(X **2)))  #we also use regularization to prevent the overfitting of the model on the training dataset
    J = J + regularization
    return J
```

**Calculation of cost function using tensorflow api in the vectorized form**


```python
def cofi_cost_func_v(X, W, b, Y, R, lambda_):
    j = (tf.linalg.matmul(X, tf.transpose(W)) + b - Y)*R  #here we are multiplying by R so that we only calculate the cost function of the movies that are rated
    J = 0.5 * tf.reduce_sum(j**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return J
num_users_r = 4
num_movies_r = 5 
num_features_r = 3

X_r = X[:num_movies_r, :num_features_r]
W_r = W[:num_users_r,  :num_features_r]
b_r = b[0, :num_users_r].reshape(1,-1)
Y_r = Y[:num_movies_r, :num_users_r]
R_r = R[:num_movies_r, :num_users_r]

J = cofi_cost_func_v(X_r, W_r, b_r, Y_r, R_r, 0);
print(f"Cost: {J:0.2f}")    
```

**New ratings done on the movie**

```python
movieList, movieList_df = load_Movie_List_pd()
my_ratings = np.zeros(num_movies)
my_ratings[2700] = 5 
my_ratings[2609] = 2;
my_ratings[929]  = 5   # Lord of the Rings: The Return of the King, The
my_ratings[246]  = 5   # Shrek (2001)
my_ratings[2716] = 3   # Inception
my_ratings[1150] = 5   # Incredibles, The (2004)
my_ratings[382]  = 2   # Amelie (Fabuleux destin d'Amélie Poulain, Le)
my_ratings[366]  = 5   # Harry Potter and the Sorcerer's Stone (a.k.a. Harry Potter and the Philosopher's Stone) (2001)
my_ratings[622]  = 5   # Harry Potter and the Chamber of Secrets (2002)
my_ratings[988]  = 3   # Eternal Sunshine of the Spotless Mind (2004)
my_ratings[2925] = 1   # Louis Theroux: Law & Disorder (2008)
my_ratings[2937] = 1   # Nothing to Declare (Rien à déclarer)
my_ratings[793]  = 5   # Pirates of the Caribbean: The Curse of the Black Pearl (2003)
for i in range(len(my_ratings)):
    if my_ratings[i]>0:
        print(f'Rating given for movie {movieList_df.loc[i,"title"]} is: {my_ratings[i]}')  #here we are extracting the title of a movie using the id
rated_movies = [i for i in range(len(my_ratings)) if my_ratings[i]>0]      
```

**Addition of those new ratings on the ratings matrix and also creating new mask based on these new ratings to show whether the movies are rated or not**

```python
Y = np.c_[my_ratings,Y]   #ratings after addition of new ratings done 
R = np.c_[(my_ratings!=0).astype(int),R] #new mask based on the addition of those new ratings
Ynorm, Ymean = normalizeRatings(Y, R)  #normalizing hte rating value
```

**Initialization of initial W,X and B**

```python
num_movies, num_users = Y.shape
num_features = 100
tf.random.set_seed(1234) # for consistent results
#initial values of W,X and b
W = tf.Variable(tf.random.normal((num_users,  num_features),dtype=tf.float64),  name='W')  #this value is different for different users
X = tf.Variable(tf.random.normal((num_movies, num_features),dtype=tf.float64),  name='X')  #this value is different for different movies
b = tf.Variable(tf.random.normal((1,          num_users),   dtype=tf.float64),  name='b')
#optimization for the parameters
optimizer = keras.optimizers.Adam(learning_rate=1e-1)  
```

**Training the collaborative filtering model**

```python
num_iterations = 200
lambda_ = 1
for i in range(num_iterations):
    with tf.GradientTape() as tape:
        cost_value = cofi_cost_func_v(X, W, b, Ynorm, R, lambda_)  #calculation of the cost function
    grads = tape.gradient(cost_value,[X,W,b])  #calculation of the partial derivative of the cost function with respect to X,W and b
    optimizer.apply_gradients(zip(grads,[X,W,b]))     #then we update the parameters X,W, and b using the partial derivative and optimizer
    if i % 20 == 0:
        print(f"Training loss at iteration {i}: {cost_value:0.1f}")
        
        
    
```

**Making the prediction of ratings**

```python
predictions = np.matmul(X.numpy(),np.transpose(W.numpy())) + b.numpy()
original_predictions = predictions + Ymean  #denormalizing the rating value
our_predictions = original_predictions[:,0]  #as  we have added ourselves in the first column among the users columns
sorted_movie_id = tf.argsort(our_predictions,direction = 'DESCENDING')  #this gives us the index or the id of the movies based on the rating value from higher to lower
for i in range(10):
    j = sorted_movie_id[i]  #extraction of the index based on the rating
    if j not in rated_movies:  #if the current j indexed movie is not rated, then we make our prediction using the model's final parameters
        print(f'The prediction of rating on movie {movieList[j]} is {our_predictions[j]}')
```
