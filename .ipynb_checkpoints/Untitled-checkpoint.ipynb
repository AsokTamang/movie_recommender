```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from utils import *
```

**Loading the dataset**

```python
X, W, b, num_movies, num_features, num_users = load_precalc_params_small()
Y, R = load_ratings_small() #here Y is the actual rating value of the movie rated by the user, R represents whether the movie is rated or not rated

print("Y", Y.shape, "R", R.shape)
print("X", X.shape)
print("W", W.shape)
print("b", b.shape)
print("num_features", num_features)
print("num_movies",   num_movies)
print("num_users",    num_users)
```

**Calculation of cost function from scratch**

```python
def cofi_cost_func(X, W, b, Y, R, lambda_):
    nm, nu = Y.shape  #nm is the number of movies and nu is the number of users
    J = 0
    for j in range(nu):
        for i in range(nm):
            if R[i,j] == 1:  #if the movie is rated by the current user, then only we do the calculation of the cost function for the movie
                J += np.sum(((W[j]@X[i]) + b[0,j] - Y[i,j]) ** 2)
    J = J/2
    regularization = ((lambda_/2) * (np.sum(W**2))) +  (((lambda_/2) * np.sum(X **2)))  #we also use regularization to prevent the overfitting of the model on the training dataset
    J = J + regularization
    return J
```

**Calculation of cost function using tensorflow api in the vectorized form**


```python
def cofi_cost_func_v(X, W, b, Y, R, lambda_):
    j = (tf.linalg.matmul(X, tf.transpose(W)) + b - Y)*R
    J = 0.5 * tf.reduce_sum(j**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return J
num_users_r = 4
num_movies_r = 5 
num_features_r = 3

X_r = X[:num_movies_r, :num_features_r]
W_r = W[:num_users_r,  :num_features_r]
b_r = b[0, :num_users_r].reshape(1,-1)
Y_r = Y[:num_movies_r, :num_users_r]
R_r = R[:num_movies_r, :num_users_r]

J = cofi_cost_func_v(X_r, W_r, b_r, Y_r, R_r, 0);
print(f"Cost: {J:0.2f}")    
```
